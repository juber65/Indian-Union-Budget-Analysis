# Bar plot for budget estimates of 2023-2024 by Ministry/Department

```python
group_data['Budget Estimates2023-2024 Total'].plot(kind='bar', figsize=(10, 6))
plt.title('Budget Estimates 2023-2024 by Ministry/Department')
plt.xlabel('Ministry/Department')
plt.ylabel('Total Budget Estimates')
plt.show()
```

# Calculate the percentage change in budget estimates between 2022-2023 and 2023-2024

```python
data['Budget Change 2022-2023 to 2023-2024'] = (
    (data['Budget Estimates2023-2024 Total'] - data['Budget Estimates 2022-2023 Total']) / 
    data['Budget Estimates 2022-2023 Total']
) * 100
```

# Display the ministries with the highest increase in budget

```python
budget_increase = data[['Ministry/Department', 'Budget Change 2022-2023 to 2023-2024']].sort_values(by='Budget Change 2022-2023 to 2023-2024', ascending=False)
print(budget_increase.head(10))
```

# Top 10 ministries by budget for 2023-2024

```python
top_ministries = data[['Ministry/Department', 'Budget Estimates2023-2024 Total']].sort_values(by='Budget Estimates2023-2024 Total', ascending=False).head(10)
```

```python
top_ministries.plot(kind='barh', x='Ministry/Department', y='Budget Estimates2023-2024 Total', legend=False, figsize=(10, 6))
plt.title('Top 10 Ministries/Departments by Budget (2023-2024)')
plt.xlabel('Total Budget Estimates (in Crores)')
plt.show()
```

# revenue vs. capital expenditure for 2023-2024

```python
data.plot(kind='scatter', x='Budget Estimates2023-2024 Revenue', y='Budget Estimates2023-2024 Capital', figsize=(10, 6))
plt.title('Revenue vs Capital Expenditure (2023-2024)')
plt.xlabel('Revenue (in Crores)')
plt.ylabel('Capital Expenditure (in Crores)')
plt.show()
```

# Group by ministry and scheme, and sum the budgets

```python
scheme_analysis = data.groupby(['Ministry/Department', 'Scheme']).sum()
```

```python
scheme_analysis['Budget Estimates2023-2024 Total'].plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.title('Budget Distribution among Schemes')
plt.ylabel('')
plt.show()
```
